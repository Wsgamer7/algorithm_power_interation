```python
import numpy as np
```

```python
def schmidt(vec1, vec2):
    normal_vec1 = vec1 / np.linalg.norm(vec1)
    orth_vec2 = vec2 - np.dot(vec2, normal_vec1) / np.dot(normal_vec1, normal_vec1) * normal_vec1
    normal_vec2 = orth_vec2 / np.linalg.norm(orth_vec2)
    return normal_vec1, normal_vec2
def same(x1, x2, delta=1e-5):
    return np.linalg.norm(x1 - x2) < delta
```

```python
v1 = np.random.rand(3)
v2 = np.random.rand(3)
n1, n2 = schmidt(v1, v2)
np.dot(n1, n2)
same(1, 1.1)
```

```python
def power_iter(A, b_k0, num_iter = 100, delta = 1e-6, isPrint=True):
    b_k = np.copy(b_k0)
    for i in range(num_iter):
        y = np.matmul(A, b_k)
        b_k1 = y / np.linalg.norm(y)
        if (np.linalg.norm(b_k1 - b_k) < delta ) or (np.linalg.norm(b_k1 + b_k)) < delta:
            break;
        b_k = b_k1
        lambda_k = np.dot(b_k1, np.dot(A, b_k1)) / np.dot(b_k1, b_k1)
        if isPrint:
            print("the {0}th iter, lambda:{1}, enginvector:{2}".format(i, lambda_k, b_k1))
    
    lambda_k = np.matmul(b_k1, np.dot(A, b_k1))
    return lambda_k, b_k1

def get_all_lambda(A, num_iter = 100, delta = 1e-6):
    lambda_lst = []
    evec_list = []
    neg_flag = False
    for i in range(A.shape[0]):
        if neg_flag:
            neg_flag = False
            continue
        b_k0 = np.random.rand(A.shape[1])
        b_k1 = np.random.rand(A.shape[1])
        print("正在计算第{0} 个特征向量".format(i))
        lambda_, evec = power_iter(A, b_k0, num_iter, delta)
        lambda2, evec2 = power_iter(A, b_k1, num_iter, delta, isPrint=False)
        if (np.abs(lambda_ - lambda2) > 1e-2):
            neg_flag = True
            print("检测到互反的特征值,正在计算第{0}和{1} 个特征向量".format(i, i+1))
            print("计算特征向量之和 :")
            lambda_sqrt, evec_added = power_iter(np.matmul(A, A), b_k0, num_iter, delta)
            print("计算特征向量之差 :")
            lambda_sqrt, evec_minus = power_iter(np.matmul(A, A), np.matmul(A, b_k0), num_iter, delta)
            evec = evec_added + evec_minus
            evec2 = evec_added - evec_minus
            # lambda_ = np.sqrt(lambda_sqrt)
            lambda_ = np.matmul(evec, np.matmul(A, evec)) / np.dot(evec, evec)
            # assert(same(np.matmul(A, evec), lambda_ * evec))
            lambda2 = -lambda_
            evec, evec2 = schmidt(evec, evec2)

        lambda_lst.append(lambda_)
        evec_list.append(evec)
        A = A - lambda_ * np.outer(evec, evec)
        #append next lambda2 and evec2 if lambda1 = - lambda2
        if (neg_flag):
            A = A - lambda2 * np.outer(evec2, evec2)
            lambda_lst.append(lambda2)
            evec_list.append(evec2)
            
    print("特征值， 特征向量")
    for i in range(len(lambda_lst)):
        print(lambda_lst[i], evec_list[i])
```

```python
A = np.array([[3, 0, 0],
              [0, 2, 0],
              [0, 0, 1]])
get_all_lambda(A, delta=1e-10)
np.linalg.eig(A)
```
